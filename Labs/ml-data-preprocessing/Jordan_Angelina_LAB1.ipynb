# Cereal Data Preprocessing and Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("/Users/angelinajordan/Desktop/Intro ML and AI/Assignments/ml-data-preprocessing/M1 & M4 - Cereals - Cereals.csv")
# Display the first 5 rows
df.head()
```

```python
df.info()
```

```python
# Compute statistics individually for numerical columns
Newdf = df.select_dtypes(include=[np.number])

mean = Newdf.mean()
median = Newdf.median()
minimum = Newdf.min()
maximum = Newdf.max()
std_dev = Newdf.std()

# Combine them into a single dataframe for clean display
summary_stats = pd.DataFrame({
    'Mean': mean,
    'Median': median,
    'Min': minimum,
    'Max': maximum,
    'Std Dev': std_dev
})

summary_stats
```

```python
#creating a correlation table
corrDF = Newdf.corr()

print(corrDF)
```
